## 分析数据

创建一个dataframe lr_house_price用于进行数据分析

```python
lr_house_price = cleaned_house_price.copy()
lr_house_price
```

由于数据中存在分类变量，无法直接建立线性回归模型，下面引入虚拟变量

```python
lr_house_price = pd.get_dummies(lr_house_price,drop_first=True,columns=['mainroad','guestroom','basement',
                                                                        'hotwaterheating','airconditioning',
                                                                        'prefarea','furnishingstatus'],dtype=int)
lr_house_price
```

划分自变量和因变量

```python
y = lr_house_price['price']
X = lr_house_price.drop('price',axis=1)
```

检查自变量之间的相关性：

```python
X.corr().abs() >0.8
```

从以上输出来看，不同变量之间相关系数的绝对值均小于或等于0.8。

给模型的线性方程添加截距：

```python
X = sm.add_constant(X)
X
```

调用OLS函数，利用最小二乘法来得到线性回归模型的参数值。

```python
model = sm.OLS(y,X)
results = model.fit()
```

```python
results.summary()
```

当我们把显著区间设定为0.05时，以上结果的P值可以看出，模型认为以下因素对房屋价格没有显著性影响：卧室数、是否为简装房。此外，常数的P值也很大，说明也没有显著影响。

可以把这些变量移除后，再次建立线性回归模型。

```python
X = X.drop(['const','bedrooms','furnishingstatus_semi-furnished'],axis=1)
X
```

```python
results = sm.OLS(y,X).fit()
results.summary()
```

可以看到，当我们把P值较大的自变量从线性回归模型中移除后，R方的值从0.682增长到了0.957，提高了模型对现有数据的拟合度。

根据各个自变量在线性回归方程中的系数来看，模型预测以下因素的增加（或存在）会显著增加房屋价格：房屋面积、厕所数、楼层数、车库容量、位于主路、有客房、有地下室、有热水器、有空调、位于城市首选社区。

线性回归模型预测以下因素的增加（或存在）会显著降低房屋价格：房屋未经装修，为毛坯房。

```python
##预测房屋信息：面积为6500平方英尺，有4个卧室、2个厕所，总共2层，不位于主路，无客人房，带地下室，有热水器，没有空调，车位数为2，位于城市首选社区，简装修。
```

```python
predict_house_price = pd.DataFrame({'area':[6500],'bedrooms':[4],'bathrooms':[2],
                                    'stories':[2],'mainroad':['yes'],'guestroom':['no'],
                                    'basement':['yes'],'hotwaterheating':['yes'],
                                    'airconditioning':['no'],'parking':[2],'prefarea':['yes'],
                                    'furnishingstatus':['semi-furnished']})
predict_house_price                                    
```

我们需要把分类变量的类型转换为Category，并且通过`categories`参数，让程序知道所有可能的分类值。这样做的原因是，预测数据包含的分类可能不全。我们需要确保引入虚拟变量的时候，不会漏掉某个或某些分类。

```python
predict_house_price['mainroad'] = pd.Categorical(predict_house_price['mainroad'],categories=['no','yes'])
predict_house_price['guestroom'] = pd.Categorical(predict_house_price['guestroom'],categories=['no','yes'])
predict_house_price['basement'] = pd.Categorical(predict_house_price['basement'],categories=['no','yes'])
predict_house_price['hotwaterheating'] = pd.Categorical(predict_house_price['hotwaterheating'],categories=['no','yes'])
predict_house_price['airconditioning'] = pd.Categorical(predict_house_price['airconditioning'],categories=['no','yes'])
predict_house_price['prefarea'] = pd.Categorical(predict_house_price['prefarea'],categories=['no','yes'])
predict_house_price['furnishingstatus'] = pd.Categorical(predict_house_price['furnishingstatus'],categories=['furnished','semi-furnished','unfurnished'])
```

给预测的分类数据引入虚拟变量

```python
predict_house_price = pd.get_dummies(predict_house_price,drop_first=True,dtype=int)
predict_house_price
```

```python
predict_house_price = predict_house_price.drop(['bedrooms','furnishingstatus_semi-furnished'],axis=1)
predict_house_price
```

接下来就可以调用线性回归模型的predict方法，获得预测价格。

```python
predicted_value = results.predict(predict_house_price)
predicted_value
```
